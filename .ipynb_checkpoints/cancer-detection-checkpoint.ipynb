### The Data

Microscopic images of lymph node tissue with a resolution of 96x96 pixels has been provided. The dataset contains a total of 220025 training images and 57458 testing images. Each image has been labelled as 0 (negative for cancer cells) or 1 (positive for cancer cells). A positive label indicates that the centre 32x32 pixels region of a patch contains at least one pixel of tumor tissue. Tumor tissue in the outer region of the patch does not influence the label.

### Loading Images

Interactive kernels with GPU currently offer about 14 GB of RAM, therefore images will be kept in the uint8 format to reduce memory usage. In the unint8 format, pixel values are integers between 0 and 255.

```python
# function for loading images using the data in our dataframe

def load_images(num_images, df):
    
    # creates a numpy array filled with zeroes 
    # creating array for number of images, number of pixels (96x96) 
    # and number of channels (red, green, blue)
    X = np.zeros([num_images, 96, 96, 3], dtype=np.uint8)
    
    # converting labels to a numpy array
    # squeeze function removes single-dimensional entries 
    # from the shape of the array
    y = np.squeeze(df.label.values)[0:num_images]
    
    # read in images using tdqm notebook to show progress
    for i, row in tqdm_notebook(df.iterrows(), total=num_images):
        if i == num_images:
            break
        X[i] = cv2.imread(row['path'])
        
    return X, y
            
    
```

```python
# load images

num_images = 6
X, y = load_images(num_images=num_images, df=df)
```
